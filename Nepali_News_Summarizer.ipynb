```python
# prompt: from df get the average number of words in summary and text

# Function to count words in a string
def count_words(text):
  if isinstance(text, str):
    return len(text.split())
  return 0

# Apply the function to create new columns for word counts
df['summary_word_count'] = df['summary'].apply(count_words)
df['text_word_count'] = df['text'].apply(count_words)

# Calculate the average word count for summary and text
average_summary_words = df['summary_word_count'].mean()
average_text_words = df['text_word_count'].mean()

print(f"Average number of words in summary: {average_summary_words}")
print(f"Average number of words in text: {average_text_words}")
```

```python
# prompt: write a simple encoder-decoder architecture with LSTM that takes input as text and generated corresponding summary

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Assuming filtered_df is available from the previous code block
# We need to use the 'text' as input and 'summary' as target

# Drop rows with NaN values in 'text' or 'summary'
filtered_df.dropna(subset=['text', 'summary'], inplace=True)

# Get the text and summary columns
texts = filtered_df['text'].tolist()
summaries = filtered_df['summary'].tolist()

# Add start and end tokens to summaries for sequence generation
summaries = ['_START_ ' + summary + ' _END_' for summary in summaries]

# --- Tokenization and Padding ---

# Parameters for tokenization
max_words = 5000  # Maximum number of words to keep in the vocabulary
max_text_len = 1000  # Maximum length of input text sequences
max_summary_len = 50 # Maximum length of output summary sequences

# Tokenize the text data
text_tokenizer = Tokenizer(num_words=max_words)
text_tokenizer.fit_on_texts(texts)
text_sequences = text_tokenizer.texts_to_sequences(texts)

# Tokenize the summary data
summary_tokenizer = Tokenizer(num_words=max_words, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
summary_tokenizer.fit_on_texts(summaries)
summary_sequences = summary_tokenizer.texts_to_sequences(summaries)

# Get vocabulary sizes
text_vocab_size = len(text_tokenizer.word_index) + 1
summary_vocab_size = len(summary_tokenizer.word_index) + 1

# Pad the sequences
encoder_input_data = pad_sequences(text_sequences, maxlen=max_text_len, padding='post')
decoder_input_data = pad_sequences(summary_sequences, maxlen=max_summary_len, padding='post')

# Prepare target sequences for the decoder (one-hot encoded)
# The target sequence for the decoder is the summary shifted by one timestep
decoder_target_data = np.zeros((len(summaries), max_summary_len, summary_vocab_size), dtype='float32')

for i, seq in enumerate(decoder_input_data):
  for t, word_index in enumerate(seq):
    if t > 0:
      decoder_target_data[i, t - 1, word_index] = 1.

# --- Define the Encoder-Decoder Model ---

# Encoder
latent_dim = 256  # Latent dimensionality of the encoding space

encoder_inputs = Input(shape=(max_text_len,))
encoder_embedding = Embedding(text_vocab_size, latent_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
# We discard `encoder_outputs` and only keep the states.
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(max_summary_len,))
decoder_embedding = Embedding(summary_vocab_size, latent_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(summary_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='rmsprop', loss='categorical_crossentropy')

# --- Train the Model ---

# You'll need to split your data into training and validation sets
# For simplicity, we'll use the whole dataset for training in this example
# In a real scenario, use train_test_split from sklearn

batch_size = 64
epochs = 50 # You might need to train for more epochs

history = model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_split=0.2) # Use 20% of data for validation

# --- Define the Inference Models ---

# Encoder model (for encoding the input text)
encoder_model = Model(encoder_inputs, encoder_states)

# Decoder model (for generating the summary)
decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_inputs_single_timestep = Input(shape=(1,))
decoder_embedding_inference = Embedding(summary_vocab_size, latent_dim)(decoder_inputs_single_timestep)

decoder_outputs_inference, state_h_inference, state_c_inference = decoder_lstm(
    decoder_embedding_inference, initial_state=decoder_states_inputs)
decoder_states_inference = [state_h_inference, state_c_inference]
decoder_outputs_inference = decoder_dense(decoder_outputs_inference)

decoder_model = Model(
    [decoder_inputs_single_timestep] + decoder_states_inputs,
    [decoder_outputs_inference] + decoder_states_inference)

# --- Inference Function ---

def decode_sequence(input_seq):
  # Encode the input as state vectors.
  states_value = encoder_model.predict(input_seq)

  # Generate empty target sequence of length 1.
  target_seq = np.zeros((1, 1))
  # Populate the first character of target sequence with the start token.
  start_token_index = summary_tokenizer.word_index['_START_']
  end_token_index = summary_tokenizer.word_index['_END_']
  target_seq[0, 0] = start_token_index

  # Sampling loop for a batch of sequences
  # (to simplify, here we assume a batch of size 1).
  stop_condition = False
  decoded_sentence = ''
  while not stop_condition:
    output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

    # Sample a token
    sampled_token_index = np.argmax(output_tokens[0, -1, :])
    sampled_word = summary_tokenizer.index_word.get(sampled_token_index, '')

    if sampled_word == '_END_':
      stop_condition = True
    else:
      decoded_sentence += ' ' + sampled_word

    # Exit condition: either hit max length
    # or find stop token.
    if len(decoded_sentence.split()) > max_summary_len or sampled_word == '_END_':
      stop_condition = True

    # Update the target sequence (of length 1).
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = sampled_token_index

    # Update states
    states_value = [h, c]

  return decoded_sentence.strip()

# --- Example Usage ---

# Take a sample text from the filtered data
sample_text_index = 0 # You can change this index
input_text = texts[sample_text_index]

# Prepare the input text for inference
input_seq = text_tokenizer.texts_to_sequences([input_text])
input_seq = pad_sequences(input_seq, maxlen=max_text_len, padding='post')

# Generate the summary
generated_summary = decode_sequence(input_seq)

print("Original Text:")
print(input_text)
print("\nGenerated Summary:")
generated_summary

# You can try this for other samples as well
# sample_text_index = 1
# input_text = texts[sample_text_index]
# input_seq = text_tokenizer.texts_to_sequences([input_text])
# input_seq = pad_sequences(input_seq, maxlen=max_text_len, padding='post')
# generated_summary = decode_sequence(input_seq)
# print("\nOriginal Text:")
# print(input_text)
# print("\nGenerated Summary:")
# print(generated_summary)

```
